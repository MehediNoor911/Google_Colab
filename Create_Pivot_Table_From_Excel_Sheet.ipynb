```python
import pandas as pd

# Load the uploaded Excel file to inspect its contents
file_path = '/content/drive/MyDrive/data/pivot_table.xlsx'
excel_data = pd.ExcelFile(file_path)

# Display sheet names to understand the structure
excel_data.sheet_names
```

```python
# Load the data from the first sheet
sheet_data = excel_data.parse('Sheet1')

# Display the first few rows to understand the structure of the data
sheet_data.head()
```

```python
# Skip the introductory rows to locate the actual dataset
cleaned_data = pd.read_excel(file_path, skiprows=3)

# Display the first few rows of the cleaned data
cleaned_data.head()
```

```python
# 1. Total Sales by Region
sales_by_region = cleaned_data.groupby('Region')['Sales'].sum().reset_index()
sales_by_region
```

```python
# Create or specify a folder in your Google Drive
folder_path = '/content/drive/My Drive/pivotTable/'
```

```python
file_name_1 = 'sales_by_region.xlsx'
file_path = os.path.join(folder_path, file_name_1)
sales_by_region.to_excel(file_path, index=False)
```

```python
# 2. Branch with Highest Total Sales
sales_by_branch = cleaned_data.groupby('Branch')['Sales'].sum().reset_index()
sales_by_branch
```

```python
file_name_2 = 'sales_by_branch.xlsx'
file_path = os.path.join(folder_path, file_name_2)
sales_by_branch.to_excel(file_path, index=False)
```

```python
# 3. Units Sold by Product
units_by_product = cleaned_data.groupby('Product')['Units'].sum().reset_index()
units_by_product
```

```python
file_name_3 = 'units_by_product.xlsx'
file_path = os.path.join(folder_path, file_name_3)
units_by_product.to_excel(file_path, index=False)
```

```python
# 4. Monthly Sales Trend
cleaned_data['Month'] = pd.to_datetime(cleaned_data['Date']).dt.to_period('M')
sales_by_month = cleaned_data.groupby('Month')['Sales'].sum().reset_index()
sales_by_month
```

```python
file_name_4 = 'sales_by_month.xlsx'
file_path = os.path.join(folder_path, file_name_4)
sales_by_month.to_excel(file_path, index=False)
```

```python
sales_by_region_product = (
    cleaned_data.groupby(['Region', 'Product'])['Sales']
    .sum()
    .reset_index()
)
highest_selling_by_region = (
    sales_by_region_product.loc[sales_by_region_product.groupby('Region')['Sales'].idxmax()]
)
highest_selling_by_region
```

```python
file_name_5 = 'highest_selling_by_region.xlsx'
file_path = os.path.join(folder_path, file_name_5)
highest_selling_by_region.to_excel(file_path, index=False)
```
